## First Finding Hamming distance:

* defined as the number of different letters in the same index, given two strings of the same length

## Second Finding Patterns-Approximately that is

* Approximate Pattern Match takes a Pattern within a Text, and some integer "d"
* This function then examines every single possible Pattern within the Text and determines if the Hamming distance is within the tolerance level (d)
* It outputs a list containing all these approximate matches

## Third: Finding How many Approximate Patterns there are

* Essentially Identical to the previous function, but only finds a count, not the list of Patterns

## Fourth: "FrequentWordsWithMismatches":

* The composite function of all the previous. This function will search through a string "Text", and search for all patterns of length "k" that maximize the ApproximatePatternMatch within the tolerance range "d"
* Note that this function requires a manual construction of all possible Patterns based upon the k 

```python
import itertools

def find_Ham(stringone,stringtwo):
    count = 0
    for i,k in zip(stringone,stringtwo):
        if i!=k:
            count+=1
    return count

def SymbolToNumber(symbol):
    bases= { "A":0 , "C":1, "G":2, "T":3}
    return bases[symbol]
def NumberToSymbol(number):
    bases= {0 :"A",  1: "C", 2 :"G", 3: "T"}
    return bases[number]
def PatternToNumber(Pattern):
    if len(Pattern)==0:
        return 0
    Symbol = Pattern[len(Pattern)-1:len(Pattern)]
    Pattern = Pattern[0:len(Pattern)-1]
    return 4 * PatternToNumber(Pattern) + SymbolToNumber(Symbol)
def NumberToPattern(index,k):
    if k==1:
        return NumberToSymbol(index)
    quotient = index/4
    remainder = index % 4
    PrefixPattern = NumberToPattern(quotient,k-1)
    symbol= NumberToSymbol(remainder)
    return PrefixPattern + symbol

def ApproxPatternMatch(Pattern,Text,d):
    k=len(Pattern)
    StartingPositions=[]
    for i in range(len(Text)):
        if find_Ham(Pattern,Text[i:i+k])<= d and len(Pattern)==len(Text[i:i+k]):
            StartingPositions.append(i)
    return StartingPositions

def ApproximatePatternCount(Text,Pattern,d):
    return len(ApproxPatternMatch(Pattern,Text,d))

def FrequentWordsWithMismatches(Text,k,d):
   if k>12 or d>3:
       print "Your Inputs are off"
       return 0
   maxlist= []
   maxi=-1
   strings=[]
   #prepare all the possible kmer combinations
   for x in itertools.combinations_with_replacement("ATGC",k):
        for z in itertools.permutations("".join(x)):
            strings.append("".join(z))
   strings= set(strings);strings=list(strings)
   for i in strings:
        maxlist.append([ApproximatePatternCount(Text,i,d),i])
   del(strings)
   strings=[]
   for pairs in maxlist:
       if pairs[0]>maxi:
           maxi=pairs[0]
   for pairs in maxlist:
       if pairs[0]==maxi:
           strings.append(pairs[1])
   strings= set(strings)
   return strings
   
def FasterFrequentWordsWithMismatches(Text,k,d):
   if k>12 or d>3:
       print "Your Inputs are off"
       return 0
   FrequentPatterns=[]
   Close=[0]*4**k
   FrequencyArray=[0]*4**k
   for i in range(len(Text)-k):
        Neighborhood=Neighbors(Text[i:i+k],d)
        for e in Neighborhood:
            index = PatternToNumber(e)
            Close[index]=1
   for i in range(4**k-1):
        if Close[i]==1:
            Pattern=NumberToPattern(i,k)
            FrequencyArray[i]=ApproximatePatternCount(Text,Pattern,d)
   maxcount=max(FrequencyArray)
   for i in range(4**k-1):
        if FrequencyArray[i]==maxcount:
            FrequentPatterns.append(NumberToPattern(i,k))
   return FrequentPatterns
    
def find_complement(string):
    stg=string[::-1]
    result=[]
    for i in stg:
        if i=="A":
            result.append("T")
        elif i=="T":
            result.append("A")
        elif i=="G":
            result.append("C")
        elif i=="C":
            result.append("G")
    return result
    
def FrequentWordswithMismatchesandComplements(Text,k,d):
    Answers=FasterFrequentWordsWithMismatches(Text,k,d)
    for i in Answers:
        Answers.append(find_complement(i))
    return Answers
        
                    
# answer = FrequentWordsWithMismatches("ACGTTGCATGTCGCATGATGCATGAGAGCT",4,1) ## IT WORKS!
# Also Correct answer= FasterFrequentWordsWithMismatches("TACTACAGTAGATAGTATACAGTACAACAACAATACCAAGATAGTAATTAATTAGATGATATTATACATTAGATGATCAAATTAATTAAGTAGATAGTAGATAGTAAGTAGATTACGATGATCAAAGTATACGATATTAAGTATACGATCAAGATGATAGTACAAAGTAAGTAAGTACAAGATCAAAGTAATTATACCAATACATTACAACAAATTAGATATTAGATAGTAATTACAAGATCAAAGTACAAGATAGTATACTACGATAGTACAAATTATACGATAGTATACTACATTAAGTAGATATTATACTACCAAAGTATACGATATTA",8,2)
# answer= FasterFrequentWordsWithMismatches("CTTGCCGGCGCCGATTATACGATCGCGGCCGCTTGCCTTCTTTATAATGCATCGGCGCCGCGATCTTGCTATATACGTACGCTTCGCTTGCATCTTGCGCGCATTACGTACTTATCGATTACTTATCTTCGATGCCGGCCGGCATATGCCGCTTTAGCATCGATCGATCGTACTTTACGCGTATAGCCGCTTCGCTTGCCGTACGCGATGCTAGCATATGCTAGCGCTAATTACTTAT",9,3)


# for i in answer:
#     print i, find_complement(i)

print ApproximatePatternCount("TACGCATTACAAAGCACA","AA",1)
```

## Let's not use all the possible combinations

* In order to avoid creating every possible permutation of k length to look for approximate matches, we'll narrow the possible combinations by creating only those combinations that could *possibly* be a "Neighbor" of the Pattern of Interest, given the tolerance level "d"

### ImmediateNeighbors:

* All possible strings similar to the Input Patter, within a tolerance level of one

### Neighbors:

* All possible strings similar to the Input Pattern, with a tolerance level of d
* uses recursion to break apart string processing each 

```python
#ImmediateNeighbors(Pattern) 
#Input = Pattern
#Output = all possible strings similar that are within the tolerance level 1

def immediateNeighbors(Pattern):
    Neighborhood = [Pattern]
    bases = {0: "A", 1:"T", 2:"G", 3:"C"}
    for i in range(len(Pattern)):
        symbol = Pattern[i]
        for z in bases.values():
            if symbol != z:
                Neighbor=Pattern[0:i]+z+Pattern[i+1:len(Pattern)]
                Neighborhood.append(Neighbor)
    return Neighborhood

#Neighbors(Pattern,d)
#Input = Pattern (string), d (integer)
#Output = All possible strings similar to Pattern within tolerance level d

def Neighbors(Pattern,d):
    if d==0:
        return Pattern
    bases = {0:"A",1:"T", 2:"G", 3:"C"}
    if len(Pattern)==1:
        return "ACGT"
    Neighborhood= []
    SuffixNeighbors = Neighbors(Pattern[1:len(Pattern)],d)
    for i in SuffixNeighbors:
        if find_Ham(Pattern[1:len(Pattern)],i) < d:
            for z in bases.values():
                Neighborhood.append(z+i)
        else:
            Neighborhood.append(Pattern[0]+i)
    return Neighborhood

# Below code for in quiz assignment 
# f=open("NeighborsOutput.txt","w")
# for i in Neighbors("ACTTGCCAAT",2):
#     f.write(str(i) + "\n")
# f.close()

#Modified neighbors produces strings exactly at the tolerance level
def ModifiedNeighbors(Pattern,d):
    if d==0:
        return Pattern
    bases = {0:"A",1:"T", 2:"G", 3:"C"}
    if len(Pattern)==1:
        return "ACGT"
    Neighborhood= []
    SuffixNeighbors = ModifiedNeighbors(Pattern[1:len(Pattern)],d)
    for i in SuffixNeighbors:
        if find_Ham(Pattern[1:len(Pattern)],i) == d:
            for z in bases.values():
                Neighborhood.append(z+i)
        else:
            Neighborhood.append(Pattern[0]+i)
    return Neighborhood


for i in  ModifiedNeighbors("ACG",1):
    print i,
```
