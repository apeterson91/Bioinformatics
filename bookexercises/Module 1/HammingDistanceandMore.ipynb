## Let's not use all the possible combinations

* In order to avoid creating every possible permutation of k length to look for approximate matches, we'll narrow the possible combinations by creating only those combinations that could *possibly* be a "Neighbor" of the Pattern of Interest, given the tolerance level "d"

### ImmediateNeighbors:

* All possible strings similar to the Input Patter, within a tolerance level of one

### Neighbors:

* All possible strings similar to the Input Pattern, with a tolerance level of d
* uses recursion to break apart string processing each 

```python
#ImmediateNeighbors(Pattern) 
#Input = Pattern
#Output = all possible strings similar that are within the tolerance level 1

def immediateNeighbors(Pattern):
    Neighborhood = [Pattern]
    bases = {0: "A", 1:"T", 2:"G", 3:"C"}
    for i in range(len(Pattern)):
        symbol = Pattern[i]
        for z in bases.values():
            if symbol != z:
                Neighbor=Pattern[0:i]+z+Pattern[i+1:len(Pattern)]
                Neighborhood.append(Neighbor)
    return Neighborhood

#Neighbors(Pattern,d)
#Input = Pattern (string), d (integer)
#Output = All possible strings similar to Pattern within tolerance level d

def Neighbors(Pattern,d):
    if d==0:
        return Pattern
    bases = {0:"A",1:"T", 2:"G", 3:"C"}
    if len(Pattern)==1:
        return "ACGT"
    Neighborhood= []
    SuffixNeighbors = Neighbors(Pattern[1:len(Pattern)],d)
    for i in SuffixNeighbors:
        if find_Ham(Pattern[1:len(Pattern)],i) < d:
            for z in bases.values():
                Neighborhood.append(z+i)
        else:
            Neighborhood.append(Pattern[0]+i)
    return Neighborhood

# Below code for in quiz assignment 
# f=open("NeighborsOutput.txt","w")
# for i in Neighbors("ACTTGCCAAT",2):
#     f.write(str(i) + "\n")
# f.close()

#Modified neighbors produces strings exactly at the tolerance level
def ModifiedNeighbors(Pattern,d):
    if d==0:
        return Pattern
    bases = {0:"A",1:"T", 2:"G", 3:"C"}
    if len(Pattern)==1:
        return "ACGT"
    Neighborhood= []
    SuffixNeighbors = ModifiedNeighbors(Pattern[1:len(Pattern)],d)
    for i in SuffixNeighbors:
        if find_Ham(Pattern[1:len(Pattern)],i) == d:
            for z in bases.values():
                Neighborhood.append(z+i)
        else:
            Neighborhood.append(Pattern[0]+i)
    return Neighborhood


for i in  ModifiedNeighbors("ACG",1):
    print i,
```
