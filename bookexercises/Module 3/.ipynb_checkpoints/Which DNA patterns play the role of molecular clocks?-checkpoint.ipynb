```python
import itertools
import collections
import MotMod

def consume(iterator, n):
    "Advance the iterator n-steps ahead. If n is none, consume entirely."
    # Use functions that consume iterators at C speed.
    if n is None:
        # feed the entire iterator into a zero-length deque
        collections.deque(iterator, maxlen=0)
    else:
        # advance to the empty slice starting at position n
        next(itertools.islice(iterator, n, n), None)
    


def find_Ham(stringone,stringtwo):
    count = 0
    for i,k in zip(stringone,stringtwo):
        if i!=k:
            count+=1
    return count

def stringpresent(Pattern,DNA,k,d):
    for string in DNA:
        for i in range(len(string)-k):
            if find_Ham(string[i:i+k],Pattern)<=d:
                return True
    return False

def Neighbors(Pattern,d):
    if d==0:
        return Pattern
    bases = {0:"A",1:"T", 2:"G", 3:"C"}
    if len(Pattern)==1:
        return "ACGT"
    Neighborhood = []
    SuffixNeighbors = Neighbors(Pattern[1:len(Pattern)],d)
    for i in SuffixNeighbors:
        if find_Ham(Pattern[1:len(Pattern)],i) <= d:
            for z in bases.values():
                Neighborhood.append(z+i)
        else:
            Neighborhood.append(Pattern[0]+i)
    return Neighborhood

#Boolean returning function determines whether Pattern is present in every string of DNA with at most d-mismatches
def Present_in_DNA(DNA,Pattern,k,d):
    truths=0;i=0
    for string in DNA: # for every string of DNA stored in DNA list
        for i in range(len(string)-k): #for every nucleotide in the DNA string
            if find_Ham(Pattern,string[i:i+k])<=d: # if there is a kmer in that dna string with <=d mismatches
                consume(string,1) # move on to the next string rather than continuing to iterate through all the other patterns
                truths+=1 # increase the counter detailing how often the kmer is found in the collective DNA set of strings
    if truths>=len(DNA):
        return True
    else:
        return False
            

def MotifEnumeration(DNA,k,d):
    City=[]
    Patterns=[string[i:i+k] for string in DNA for i in range(len(string)-k)]
    City=[Neighbors(Pattern,d) for Pattern in Patterns] # collect all possible variations of kmers within d tolerance
    BestPatterns=[Neighborhood[i] for Neighborhood in City for i in range(len(Neighborhood)) if Present_in_DNA(DNA,Neighborhood[i],k,d)]
#     for Neighborhood in City: # for each list of possible kmers 
#         for Neighbor in Neighborhood: # for each possible kmer in each list
#             if Present_in_DNA(DNA,Neighbor,k,d): # if that kmer is present in every string of DNA with tolerance level d
#                 BestPatterns.append(Neighbor) # add that kmer to the "Best Patterns" list to return
    BestPatterns=set(BestPatterns);BestPatterns=list(BestPatterns) # remove duplicates
    return BestPatterns # return list


Example=["ATTTGGC","TGCCTTA","CGGTATC", "GAAAATT"]
answers=["ATA", "ATT", "GTT", "TTT"]

# print Present_in_DNA(Example,"TTT",4,2)

print "\n".join(MotifEnumeration(Example,3,1))
# print set(answers)==set(MotifEnumeration(Example,3,1))
```

### Scoring Motifs

Use a matrix of the varying candidate kmers to create profile of similarity

Four important concepts we'll use to build a better idea of what motif would most likely be found in as a TBF- all of these will be matrices of different types.

1. Motif Matrix:
    * Stores each potential motif kmer as a string delineated by individual nucleotides.
    * dimensions: k x n where k is the length of the string, and n is the number of candidate motifs
    * Use: This matrix can be used to understand which nucleotide is the most popular in each position of the kmer
2. Count Matrix:
    * Matrix holding the count of each nucleotide in each motif string COLUMN
    * This is a 4 x k matrix where each row holds information pertaining to a specfic nucleotide
    * Use: Used to construct the profile matrix
3. Profile Matrix:
    * Matrix holding the *frequency* of each Nucleotide in each column (i.e. position) of the nucleotide kmer
    * dimensions: 4 x k again, where each row holds information pertaining to a specific nucleotide at each position in the kmer
    * Use: Constructing the Consensus Motif
4. Consensus Motif:
    * Using the Profile matrix, we can determine what might be the most "idealized" candidate for a (k,d) motif
    


```python
import numpy as np
import math as m
#Test data
# s1="TCGGGGGTTTTT"
# s2="CCGGTGACTTAC"
# s3="ACGGGGATTTTC"
# s4="TTGGGGACTTTT"
# s5="AAGGGGACTTCC"
# s6="TTGGGGACTTCC"
# s7="TCGGGGATTCAT"
# s8="TCGGGGATTCCT"
# s9="TAGGGGAACTAC"
# s10="TCGGGTATAACC"
# sses=[s1,s2,s3,s4,s5,s6,s7,s8,s9,s10]

def MotifMat(DNA):
    return np.array([list(string) for string in DNA])

def CountMat(MotifMatrix):
    CountMatrix=np.empty([4,len(MotifMatrix[0])])
    for i in range(len(MotifMatrix[0])):
        CountMatrix[0,i]=sum(1 if x=="A" else 0 for x in MotifMatrix[:,i] )
        CountMatrix[1,i]=sum(1 if x=="C" else 0 for x in MotifMatrix[:,i])
        CountMatrix[2,i]=sum(1 if x=="G" else 0 for x in MotifMatrix[:,i])
        CountMatrix[3,i]=sum(1 if x=="T" else 0 for x in MotifMatrix[:,i])
    return CountMatrix

def CountMatPseudo(MotifMatrix):
    CountMatrix=np.empty([4,len(MotifMatrix[0])])
    for i in range(len(MotifMatrix[0])):
        CountMatrix[0,i]=sum(1 if x=="A" else 0 for x in MotifMatrix[:,i])+1
        CountMatrix[1,i]=sum(1 if x=="C" else 0 for x in MotifMatrix[:,i])+1
        CountMatrix[2,i]=sum(1 if x=="G" else 0 for x in MotifMatrix[:,i])+1
        CountMatrix[3,i]=sum(1 if x=="T" else 0 for x in MotifMatrix[:,i])+1
    return CountMatrix

def ProfileMat(CountMatrix):
    ProfileMatrix=np.empty([4,len(CountMatrix[0])])
    for i in range(len(CountMatrix[0])):
        ProfileMatrix[0,i]=(CountMatrix[0,i])/(sum(CountMatrix[:,i]))
        ProfileMatrix[1,i]=CountMatrix[1,i]/sum(CountMatrix[:,i])
        ProfileMatrix[2,i]=CountMatrix[2,i]/sum(CountMatrix[:,i])
        ProfileMatrix[3,i]=CountMatrix[3,i]/sum(CountMatrix[:,i])
    return ProfileMatrix

def Get_Consensus(ProfileMatrix):
    index = {0:"A", 1: "C", 2:"G", 3: "T"}
    consensus_string=""
    for i in ProfileMatrix.argmax(axis=0):
        consensus_string+=index[i]
    return consensus_string

def Get_Entropy(ProfileMatrix):
    runningsum=[]
    for i in range(len(ProfileMatrix[0])):
        runningsum.append(Entropy_Calculation(ProfileMatrix[:,i]))
    print runningsum
    return sum(runningsum)
        
def Entropy_Calculation(l):
    summation=0
    for i in l:
        if i==0:
            pass
        else:
            summation+= i * m.log(i,2)
    return -summation
```

### Eureka!

So instead of searching for a collection of kmers *Motifs* by minimizing $Scores(Motifs)$ we can instead search for a potential consensus string *Pattern* minimizing $d(Pattern,Motifs)$ among all possible kmers and all possible choices of kmers *Motifs* in our DNA sample

This yields our 
### Equivalent Motif Finding Problem: 
Given a collection of strings, find a collection of k-mers (one from each string) that minimizes the distance between all possible patterns and all possible collections of k-mers.
     * Input: A collection of strings Dna and an integer k.
     * Output: A k-mer Pattern and a collection of k-mers, one from each string in Dna, minimizing d(Pattern, Motifs) among all possible choices of Pattern and Motifs.

```python
def d(Pattern,Motifs):
    summation=0
    for Motif in Motifs:
        summation+=find_Ham(Pattern,Motif)
    return summation            
```

### The Median String Problem

Why doesn't this switch in approaches make things harder? Because we don't have to search **every** motif in order to minimize d(Pattern,Motifs)

To further explain this- let $Motifs(Pattern,DNA)$ Be a collection of kmers that minimizes d(Pattern,Motifs) for a given Pattern and all possible kmers *Motif* in the DNA  For example, in the problem below the five colored strings represents $Motifs("AAA",Dna)$
<img src="img_2Motifs.png">


The reason why we don’t need to consider all possible collections Motifs is that we can generate the strings in Motifs(Pattern, Dna) one at a time. Given a k-mer Pattern and a longer string Text, we use d(Pattern, Text) to denote the minimum Hamming distance between Pattern and any k-mer in Text,$$d(Pattern,Text)=min(HammingDistance(Pattern,Pattern'))$$ where we take the minimum Hamming Distance of all the Patterns in the Text.

Another definition:

$$d(Pattern,DNA)= \sum_{i=0}^{t} d(Pattern,DNA_{i})$$

Our goal is to find a k-mer Pattern that minimizes d(Pattern, Dna) over all k-mers Pattern, the same task that the Equivalent Motif Finding Problem is trying to achieve. We call such a k-mer a median string for Dna.

**Median String Problem: Find a median string.**
     Input: A collection of strings Dna and an integer k.
     Output: A k-mer Pattern that minimizes d(Pattern, Dna) among all k-mers Pattern.

Notice that finding a median string requires solving a double minimization problem. We must find a k-mer Pattern that minimizes d(Pattern, Dna), where this function is itself computed by taking a minimum over all choices of k-mers from each string in Dna.

```python
def DistanceBetweenStrings(Pattern,DNA):
    k=len(Pattern)
    distance=0
    for string in DNA:
        HammingDistance=1*(10**10)
        for i in range(len(string)-k):
            if HammingDistance>find_Ham(Pattern,string[i:i+k]):
                HammingDistance=find_Ham(Pattern,string[i:i+k])
        distance+=HammingDistance
    return distance

def MedianString(DNA,k):
    distance = 1*(10**100)
    Median=""
    for Pattern in [MotMod.NumberToPattern(i,k) for i in range(4**k)] :
        if distance > DistanceBetweenStrings(Pattern,DNA):
            distance=DistanceBetweenStrings(Pattern,DNA)
            Median=Pattern
    print distance
    return "".join(Median)

# print MedianString(MotMod.readintextdata("DBPT.txt"),6)

DNA=["CTCGATGAGTAGGAAAGTAGTTTCACTGGGCGAACCACCCCGGCGCTAATCCTAGTGCCC",
"GCAATCCTACCCGAGGCCACATATCAGTAGGAACTAGAACCACCACGGGTGGCTAGTTTC",
"GGTGTTGAACCACGGGGTTAGTTTCATCTATTGTAGGAATCGGCTTCAAATCCTACACAG"]

print DistanceBetweenStrings("TCTGAAG",DNA)
print DistanceBetweenStrings("GTCAGCG",DNA)
print DistanceBetweenStrings("GGTTACT",DNA)
print DistanceBetweenStrings("TAGTTTC",DNA)
print DistanceBetweenStrings("GATGAGT",DNA)
print DistanceBetweenStrings("CGTGTAA",DNA)
print MedianString(DNA,7)
```

### Why have we reformulated the Motif Problem?

simply put: time. The running time of the BruteForceAlgorithim was $ O(n^{t} * k * t)}$ (where t is the number of strings of DNA), and the running time of the MedianStringAlgorithim was $O(4^{k} * n * t  * k}$ which is far more favorable since k is almost always less than or equal to 20, whereas t is in the thousands.

The Median String Problem teaches an important lesson, which is that sometimes rethinking how a problem is formulated can lead to dramatic improvements in the runtime required to solve it. In this case, our simple observation that Score(Motifs) could just as easily be computed row-by-row as column-by-column produced the faster MEDIANSTRING algorithm.

Of course, the ultimate test of a bioinformatics algorithm is how it performs in practice. Unfortunately, since MEDIANSTRING has to consider 4k k-mers, it becomes too slow for the Subtle Motif Problem, for which k = 15. We will run MEDIANSTRING with k = 13 in the hope that it will capture a substring of the correct 15-mer motif. The algorithm still requires half a day to run on our computer and returns the median string AAAAAtAGaGGGG (with distance 29). This 13-mer is not a substring of the implanted pattern AAAAAAAAGGGGGGG. but it does come close.

We have thus far assumed that the value of k is known in advance, which is not the case in practice. As a result, we are forced to run our motif finding algorithms for different values of k and then try to deduce the correct motif length. Since some regulatory motifs are rather long — later in the chapter, we will search for a biologically important motif of length 20 — MEDIANSTRING may be too slow to find them.

### Randomized Motif Search

Randomized Algorithms can be surprisingly powerful tools to efficiently find answers to problems that cannot be easily deterministically derived 

```python
import random
def Motifs_from_Profile(Profile,DNA):
    k=len(Profile[0])
    Motifs=[]
    for strand in DNA:
        Motifs.append(Profile_MostProbable(strand,k,Profile))
    return Motifs

def random_kmer(string,k):
    x=len(string)
    while len(string)-x<k:
        x=random.randrange(0,len(string))
    return string[x:x+k]

def initialize_randomMotifs(k,DNA):
    Motifs=[]
    for string in DNA:
        Motifs.append(random_kmer(string,k))
    return Motifs

def RandomizedMotifSearch(DNA,k,t):
    i=0;j=0;a=0;switch=True
    while i<1000:
        random.seed(a);a+=1
        Motifs=initialize_randomMotifs(k,DNA)
        if i==0:
            BestMotifs=Motifs
        while 1==1:
            ProfileMatrix=ProfileMat(CountMatPseudo(MotifMat(Motifs)))
            Motifs=Motifs_from_Profile(ProfileMatrix,DNA)
            if MotifScore(Motifs,k)<MotifScore(BestMotifs,k):
                BestMotifs=Motifs
            else:
                break
        i+=1
    return BestMotifs



# answers=RandomizedMotifSearch(readintextdata("RMS.txt",True),15,20)
# print "\n".join(answers)
# print MotifScore(answers,15) # Works!
```

### Gibbs Sampling:

A more cautious method of finding the best motifs, using a more "advanced" Random Number generator

```python
def Select_Pseudo_RandMotif(X,ProfileMatrix,k):
    # from all the kmers in the "special strand"
    kmers=[X[i:i+k] for i in range(len(X)-k)]
    #compute the probability that that kmer is a motif dictated by the profile matrix
    probabilities=[computeprobability(kmer,ProfileMatrix) for kmer in kmers]
    BD=[x/sum(probabilities) for x in probabilities]
    randroll=random.random();summation=float(0)
    for i in range(len(BD)):
        summation+=BD[i]
        if randroll<summation:
            randomkmer=i
            break
    return kmers[randomkmer]

def GibbsSampler(DNA,k,t,N):
    Motifs=initialize_randomMotifs(k,DNA) # Select a random kmer motif in each DNA string as starter set
    BestMotifs=list(Motifs);switch=0
#     print "initial" + str(MotifScore(BestMotifs,k))
    for j in range(N):#Run N iterations
        X=random.randrange(0,len(DNA))
        ProfileMatrix=ProfileMat(CountMatPseudo(MotifMat([Motifs[i] for i in range(len(Motifs)) if i!=X])))
        Motifs[X]=Select_Pseudo_RandMotif(DNA[X],ProfileMatrix,k)
        if MotifScore(Motifs,k)<MotifScore(BestMotifs,k):
            BestMotifs=list(Motifs)
            switch+=1
    print BestMotifs,MotifScore(BestMotifs,k)
    return BestMotifs

# testing=["CGCCCCTCTCGGGGGTGTTCAGTAAACGGCCA",
#          "GGGCGAGGTATGTGTAAGTGCCAAGGTGCCAG",
#          "TAGTACCGAGACCGAAAGAAGTATACAGGCGT", 
#          "TAGATCAAGTTTCAGGTGCACGTCGGTGAACC",
#          "AATCCACCAGCTCCACGTGCAATGTTGGCCTA"]
# answersone=set(GibbsSampler(testing,8,5,1000))
# answers=set(["TCTCGGGG","CCAAGGTG","TACAGGCG","TTCAGGTG","TCCACGTG"])
# print MotifScore(answers,8)

print "\n".join(GibbsSampler(readintextdata("Gibbs.txt",True),15,20,2000))
```
